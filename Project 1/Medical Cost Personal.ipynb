# Introduction of the dataset

Insurance is a pivotal sector of the financial world, allowing individuals and businesses to protect themselves against uncertain risks. Its essence lies in the pooling of risks, where a multitude of policyholders contribute premiums to a central fund, out of which claims are paid. The task of predicting these claims and understanding the various factors that influence insurance costs is of paramount importance to insurers. With the advent of big data and sophisticated algorithms, the insurance industry is experiencing a paradigm shift, with predictive modeling playing a central role in its evolution. Linear Regression, a fundamental statistical and machine learning technique, offers insurers a powerful tool to predict and forecast insurance costs based on a range of independent variables

# Problem Statement

The insurance industry is grappling with several challenges, one of which is the accurate forecasting of policy premiums and claim amounts. With an increasing number of variables influencing insurance costs – from policyholder demographics to their health and behavior – it has become essential to employ statistical methods that can quantify these relationships. The problem thus becomes:

"How can we leverage linear regression techniques to predict insurance premiums and claims based on available data and variables?"


# Strategy

To address this, we aim to:
** Gather a dataset of insurance costs and related variables.
**  Explore the dataset and identify the key variables that influence insurance costs.
** Identify and collect relevant data points that influence insurance premiums and claims.
**  Build a linear regression model to establish the relationship between these data points and the insurance outcomes (premiums or claims).
**  Validate the accuracy and reliability of this model using various metrics and tests.
**  Use this model as a forecasting tool to aid insurers in their decision-making processes and to optimize premium pricing strategies.

In doing so, we hope to create a framework that not only predicts future insurance costs with higher accuracy but also offers insights into the primary drivers of these costs. This understanding can be leveraged by insurance companies to refine their risk assessment strategies and make more informed business decisions

# Exploring the target variable: Charges

```python
Medical_cost_personal['charges'].describe()
```

The average medical cost is 13270.42, with a standard deviation of 12110.01.

The lowest cost of medical cost is 1121.87, while the highest cost is 63770.43.

The median of medical cost is 9382.03, which is lower than the mean of 13270.42, indicating that the distribution of medical cost is right-skewed.

An histogram and boxplot  of medical cost is will help us to visualize the distribution of medical cost.

```python
def histrogram(col):

   # creating a grid of 3 columns
    fig = plt.figure(constrained_layout=True, figsize=(15, 10))
    
    grid = GridSpec(ncols=3, nrows=1, figure=fig)
    sns.displot(Medical_cost_personal[col])
    plt.title('Histogram of {}'.format(col))
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

    sns.boxplot(Medical_cost_personal[col])
    plt.title('Boxplot of {}'.format(col))
    plt.xlabel(col)
    plt.show()


histrogram('charges')
```

From the histogram, we can see that the distribution of medical cost is right-skewed, with a long tail on the right.
We can also see that most of the medical cost is below 14 000.

From the boxplot, we can see that there are a few outliers in the medical cost, with the highest cost being 63770.43.

We can therefore conclude that the distribution of medical cost is right-skewed, with a few outliers on the right.

A log transformation of medical cost will help us to normalize the distribution of medical cost.

```python
Medical_cost_personal['Charges_log'] = np.log(Medical_cost_personal['charges'])

#histrogram(np.log(Medical_cost_personal['charges']))
```

```python
plt.figure(figsize=(8, 6))
plt.title('Log Normal Distribution')
sns.distplot(y, fit=stats.lognorm, kde=False)
```

```python

# Plot the Johnson SU distribution
plt.figure(figsize=(8, 6))
plt.title('Johnson SU Distribution')
sns.distplot(y, fit=stats.johnsonsu, kde=False)
```

It is apparent the charges are not normally distributed. They are skewed right and contain outliers, as evidenced by box plot, distribtion plot and histogram. This is important to note, as several statistical tests rely on normality. We will log transform the charges to reduce the skewness. This is a common transformation for right skewed data. As both the johnsonsu distribution are both good fits.
